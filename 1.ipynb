```python
class Node(object):
    def __init__(self, value=None, next_node=None):
        self.value = value
        self.next = next_node

    def __str__(self):
        return f"[Node with value {self.value}]"
    
    def __lt__(self, other):
        return self.value < other.value
    
    def swap_values(self, other):
        self.value, other.value = other.value, self.value

def print_linked_list(head):
    cur = head
    while cur is not None:
        print(cur)
        cur = cur.next
```

```python
h, a, b, c, d = Node(1), Node(2), Node(3), Node("Внезапно"), Node(5)

h.next = a
a.next = b
b.next = c
c.next = d

print_linked_list(h)
```

```python
def reverse_linked_list(head):
    """
    На вход она принимает головную ноду,
    а на выход отдаёт хвостовую ноду исходного списка, 
    """
    previous = None
    while head is not None:
        next_ = head.next
        head.next = previous
        previous, head = head, next_
    return previous
        
```

```python
print_linked_list(h)
h = reverse_linked_list(h)
print("---")
print_linked_list(h)
```

```python
import random

h, a, b, c, d = [Node(random.randint(0, 10)) for _ in range(5)]
h.next = a
a.next = b
b.next = c
c.next = d
```

```python
def sort(head):
    while head is not None:
        next_ = head.next
        while next_ is not None:
            if head < next_:
                head.swap_values(next_)
            next_ = next_.next
        head = head.next
                
    return head
```

```python
print_linked_list(h)
sort(h)
print('sorted')
print_linked_list(h)
```
